```python
import random
from hangman_art import stages, logo
from hangman_words import word_list
from jpg_file import jpgs
import matplotlib.pyplot as plt
import os
from IPython.display import Image, display

a=os.listdir("hangman")
path=[]
for i in range(len(a)):
    path.append(os.path.join("hangman",a[i]))
list_of_images_in_folder=[]
print(logo)
game_is_finished = False
lives = len(path) - 1

chosen_word = random.choice(word_list)
word_length = len(chosen_word)

display_word = ["_"]*word_length

while not game_is_finished:
    guess = input("Guess a letter: ").lower()
    if guess in display_word:
        print(f"You've already guessed {guess}")

    for position in range(word_length):
        letter = chosen_word[position]
        if letter == guess:
            display_word[position] = letter
    print(f"{' '.join(display_word)}")

    if guess not in chosen_word:
        print(f"You guessed {guess}, that's not in the word. You lose a life.")
        lives -= 1
        
        if lives == 0:
            game_is_finished = True
            print("You lose.")
    
    if not "_" in display_word:
        game_is_finished = True
        print("You win.")

    #print(stages[lives])
    #from IPython.display import Image, display
    print(f"{lives} lives left")
    display(Image(filename=path[10-lives]))
    
```
