```python
from IPython.display import clear_output

def add(n,m):
    return n+m
def sub(n,m):
    return n-m
def div(n,m):
    return n/m
def mul(n,m):
    return n*m


def calculator():
    calc = True
    operation = {
        "+" : add,
        "-" : sub,
        "*" : mul,
        "/" : div
    }
    display_calc = ""
    print(lo+go)
    n = float(input("Enter First Number!!!\n"))
    clear_output()
    display_calc += str(n)

    while calc:
        print(lo+display_calc+go)
        oper = input("Select the Operation to perform:\n+\n-\n*\n/\n")
        clear_output()
        display_calc += oper
        print(lo+display_calc+go)
        m = float(input("Enter Second Digit!!!\n"))
        display_calc += str(m)
        clear_output()
        print(lo+display_calc+go)
        ans = operation[oper](m,n)
        clear_output(wait = True)
        print(f"{n}{oper}{m} = {ans}")
        display_calc+= "="
        display_calc+=str(ans)
        print(lo+display_calc+go)
        n = ans
        cont = input("Do you want to continue the same?Type y or n\n")
        if cont == "n":
            calc = False
            restart = input("do you want to restart?Type y or n\n")
            if restart == "y":
                clear_output()
                calculator()
        display_calc = str(ans)
calculator()
```
