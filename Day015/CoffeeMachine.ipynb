```python
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
            "milk": 0
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}
earned_money = 0
resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}
```

```python
def print_report(water, milk, coffee, money, report):
    
    print(f"Water: {water}\nMilk: {milk}\nCoffee: {coffee}")
    
    
def user_input():
    
    choice = input("What would you like to have? (espresso/latte/cappuccino)")
    
    return choice

def check_resources(Menu, choice, resources):
    
    if choice == "report":
        
        print(resources)
        
        return "Hello"
    
    
    requirements = Menu[choice]["ingredients"]
    
    
    if requirements["milk"] > resources["milk"] or requirements["water"] > resources["water"] or requirements["coffee"] > resources["coffee"]:
    
        if requirements["milk"] > resources["milk"]:
            
            print("Sorry there is not enough Milk!!!")
            
            return user_input()

        if requirements["water"] > resources["water"]:
            
            print("Sorry there is not enough Water")

            return user_input()

        if requirements["coffee"] > resources["coffee"]:
            
            print("Sorry there is not enough Coffee")
            
            return user_input()
    
    else:
        
        return insert_coin(choice, Menu)

def insert_coin(choice, Menu):
    
    print("Please Insert Coins!!")
    
    money = 0
    
    money += 25 * int(input("How many Quaters?"))
    money += 10 * int(input("How many Dimes?"))
    money += int(input("How many Nickels?"))
    money += int(input("How many Pennys?"))
    
    price = 100 * Menu[choice]["cost"]
    ingredients = Menu[choice]["ingredients"]
    
    global earned_money
    
    if price > money:
        
        print(f"That's not enough money!!, here is your refund ${money}")
        
        return 
    
    elif price == money :
        
        earned_money += price
        
        resources["water"] -= ingredients["water"]
        resources["coffee"] -= ingredients["coffee"]
        resources["milk"] -= ingredients["milk"]
        
        print(f"Here is your {choice}")
        
        return 
    
    elif price < money:
        
        earned_money += price
        
        resources["water"] -= ingredients["water"]
        resources["coffee"] -= ingredients["coffee"]
        resources["milk"] -= ingredients["milk"]
        
        print(f"Here is your {choice}")
        
        refund  = (money - price)/100
        print(f"Here is your refund ${refund}")
        
        return 
    
    else:
        
        pass
    
def machine(Menu , resources , earned_money = 0):
    
    print(f"money : {earned_money}")
    
    choice = user_input()
    
    check_resources(Menu, choice, resources)

    cont = input("Do you want to continue, press yes to continue and anything else to discontinue")
    
    if cont == "yes":
        
        return machine(Menu , resources ,earned_money)
    
    else:
        
        cont = input("Do you want to buy anything, press yes to continue and STOP to shut-down")
        
        if cont == "yes":
            
            return machine(Menu , resources)
        
        if cont == "STOP":
            
            return "Machine has shut down"
```

```python
machine(MENU , resources)
```
